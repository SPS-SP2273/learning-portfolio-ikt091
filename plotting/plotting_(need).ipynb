## 1 A simple plot

### 1.1 Letâ€™s look at some code.

```python
from matplotlib import pyplot as plt
```

```python


x = [0, 1, 2, 3, 4, 5, 6]

y_1 = [0, 2, 4, 6, 8, 10, 12]

y_2 = [0, 3, 6, 9, 12, 15, 18]

y_3 = [0, 4, 8, 12, 16, 20, 24]

x_error, y_error = .1, 0.75

plt.plot(x, y_1, color = 'red', linestyle = 'dashed', label = 'Y values')
plt.plot(x, y_2, c = 'blue', label = 'Y2 values')
plt.errorbar(x, y_3, xerr = x_error, yerr = y_error, c = 'green', label = 'Y3 with errors')
plt.xlabel('x_values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=1)
plt.legend(loc = 'upper left')


plt.show()
```

### 1.5 Better with NumPy

```python
import numpy as np
```

```python
x = [0, 1, 2, 3, 4, 5, 6]

y_1 = [0, 2, 4, 6, 8, 10, 12]

y_2 = [0, 3, 6, 9, 12, 15, 18]

y_3 = [0, 4, 8, 12, 16, 20, 24]

np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)
np_y_3 = np.array(y_3)

x_error, y_error = .1, 0.75

plt.plot(np_x, np_y_1, color = 'red', linestyle = 'dashed', label = 'Y values')
plt.plot(np_x, np_y_2, c = 'blue', label = 'Y2 values')
plt.errorbar(np_x, np_y_3, xerr = x_error, yerr = y_error, c = 'green', label = 'Y3 with errors')
plt.xlabel('x_values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=1)
plt.legend(loc = 'upper left')

plt.show()
```

### 1.6 Adding mathematical functions (also why NumPy lists are better)

## Exercise 2 :  A simple mathematical plot

```python
import math

exponent = 4
print ("Exponential Value is: ", math.exp(exponent))
```

```python

x_list = []
y_list = []

for x in range(50): # goes with integer, so curve not smooth
    x_list += [x]
    y1 = np.sin(x)
    y2 = math.exp(-x/10) # math cannot handle array 
    y = y1*y2
    y_list += [y]

x_array = np.array(x_list)
y_array = np.array(y_list)
print(x_array)
print(y_array)
x = np.linspace(0,50,1000)
plt.ylim([-1, 1])
plt.plot(x_array, y_array)
plt.xlabel('x', fontsize = 15, c = 'darkblue')
plt.ylabel('y', fontsize = 15, c = 'darkblue')
plt.title(r'$y = (e^{-x/10})\mathrm{sin}(x)$', fontsize = 20)
plt.grid(alpha=.25)

plt.show()
```

```python
# CORRECT CODE
# The correct way!

x = np.linspace(0,50,1000)
y = np.exp(-x/10)*(np.sin(x))


plt.ylim([-1, 1])
plt.plot(x, y)
plt.xlabel('x', fontsize = 15, c = 'darkblue')
plt.ylabel('y', fontsize = 15, c = 'darkblue')
plt.title(r'$y = (e^{-x/10})\mathrm{sin}(x)$', fontsize = 20)
plt.grid(alpha=.25)

plt.show()
```
