## 1 A Bucket of water

### 1.1 Getting a feel

```python
import numpy as np
import os
import glob
import shutil
import math
from matplotlib import pyplot as plt
import numpy.random as rnd
```

```python
max_time = 5 # Max. time (min)
dt = .5 # Time step (min)
rate = 5 # Rate (L/min)
all_volume = [] # Keeps track of all volumes
volume = 0 # Starting vol

all_time = np.arange(start = 0, stop = max_time, step = dt)

for time in all_time:
    all_volume.append(volume) # Record vol
    dV = rate * dt # Calculate change in vol
    volume += dV # Update the new volume
    
plt.plot(all_time, all_volume)
plt.ylabel('Volume(L)')
plt.xlabel('Time(mins)')
```

### 1.2 An overflowing bucket

```python
max_time = 5 # Max. time (min)
dt = .5 # Time step (min)
rate = 5 # Rate (L/min)
all_volume = [] # Keeps track of all volumes
bucket_capacity = 9
volume = 0 # Starting vol

all_time = np.arange(start = 0, stop = max_time, step = dt)

for time in all_time:
    all_volume.append(volume) # Record vol
    dV = rate * dt # Calculate change in vol
    if volume <= bucket_capacity:
        volume += dV # Update the new volume
        
plt.plot(all_time, all_volume)
plt.ylabel('Volume(L)')
plt.xlabel('Time(mins)')
plt.hlines(bucket_capacity, 0, max_time, color = 'grey', ls = 'dashed')
```

### 1.3 A leaky bucket

```python
max_time = 5
dt = .001 # going with the one selected earlier
rate = 5
leak_rate = 1.5
bucket_capacity = 9 
all_vol = []
vol = 0

all_time = np.arange(start = 0, stop = max_time, step = dt)

for time in all_time:
    all_vol.append(vol)
    
    dV = rate * dt
    leak_vol = leak_rate * dt
    vol -= leak_vol
    
    if vol <= bucket_capacity:
        vol += dV
        
plt.plot(all_time, all_vol, label = 'volume of water')
plt.ylabel('Vol(L)')
plt.ylim([0, 10])
plt.xlim([0, 5])
plt.xlabel('Time(mins)')
plt.hlines(bucket_capacity, 0, max_time, color = 'grey', ls = 'dashed')
plt.legend()

plt.show()
```

### 1.4 Lets turn off the tap

```python
max_time = 5
dt = .001 # going with the one selected earlier
rate = 5
leak_rate = 1.5
bucket_capacity = 9 
all_vol = []
vol = 0
tap_off_time = 3

all_time = np.arange(start = 0, stop = max_time, step = dt)

for time in all_time:
    all_vol.append(vol)
    
    dV = rate * dt
    leak_vol = leak_rate * dt
    vol -= leak_vol
    
    if (vol <= bucket_capacity) and (time < tap_off_time):
        vol += dV

plt.plot(all_time, all_vol)
plt.ylabel('Vol (L)')
plt.xlabel('Time(min)')
plt.hlines(bucket_capacity, 0, max_time, colors = 'grey', ls = 'dashed')

plt.show()
```

```python
decay_constant = 142
stop_fraction = 1E-3 # stop when the sample has shrunk to this fraction of the starting value

N0 = 1
dt = .001
time, N = 0, N0 # time = 0, N = N0

all_N, all_time = [], [] # all_N = [], all_time = []

while True:
    all_time.append(time)
    all_N.append(N)
    
    dN = -decay_constant*N*dt
    N += dN
    
    if N < N0*stop_fraction:
        break
        
    time += dt
    
plt.plot(all_time, all_N)
plt.ylabel('$N$')
plt.xlabel('Time (Millions of years)')

plt.show()
```

## Exercise 1 :  A more realistic leak

```python
def leak_rate(vol):
    h = vol*10**(-3)/.07
    leak_rate = ((1e-4*np.sqrt(2*9.8*h))/(10**(-3)))*60
    return leak_rate

max_time = 5
dt = .01 
rate = 5
bucket_capacity = 9 
all_vol = []
vol = 0
all_time = []
tap_off_time = 3
time = 0

while time <= max_time:
    all_time.append(time)
    all_vol.append(vol)
    
    dV = rate * dt
    leak_vol = leak_rate(vol) * dt
    vol -= leak_vol
    
    if (vol <= bucket_capacity) and (time < tap_off_time):
        vol += dV
        
    if vol < 0:
        vol = 0
    
    time += dt


plt.plot(all_time, all_vol)
plt.ylabel('Vol (L)')
plt.xlabel('Time(mins)')
plt.xlim([0,4])

plt.show()
```
